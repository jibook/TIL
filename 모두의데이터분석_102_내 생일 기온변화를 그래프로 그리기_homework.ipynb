```python
import csv
import matplotlib.pyplot as plt

f = open("C:/Users/jhjh3/Desktop/jicode/data/seoul_book.csv")
data=csv.reader(f)
next(data)
high_book=[]
low_book=[]


for row in data:
    if row[-1] != '' and row[-2] != '':
        if 1983 <= int(row[0].split('-')[0]) :
            if row[0].split('-')[1]=='03' and row[0].split('-')[2]=='06':
                high_book.append(float(row[-1]))
                low_book.append(float(row[-2]))

                       
plt.plot(high_book, 'pink')
plt.plot(low_book, 'skyblue')
plt.show()
```

```python
plt.rc('font',family="Malgun Gothic")
plt.rcParams['axes.unicode_minus']=False # 한글폰트 사용시 - 부호 표현
plt.title("지현쓰 생일 기온 변화 그래프")

                       
plt.plot(high_book, 'pink')
plt.plot(low_book, 'skyblue')
plt.show()
```

### 내 생일 기온 변화를 그래프로 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open("C:/Users/jhjh3/Desktop/jicode/data/seoul_book.csv")
data=csv.reader(f)
next(data)
high_book=[]
low_book=[]


for row in data:
    if row[-1] != '' and row[-2] != '':
        if 1996 <= int(row[0].split('-')[0]) :
            if row[0].split('-')[1]=='03' and row[0].split('-')[2]=='06':
                high_book.append(float(row[-1]))
                low_book.append(float(row[-2]))


plt.rc('font',family="Malgun Gothic")
plt.rcParams['axes.unicode_minus']=False # 한글폰트 사용시 - 부호 표현
plt.title("지현쓰 생일 기온 변화 그래프")
                       
plt.plot(high_book, 'hotpink', label='high')
plt.plot(low_book, 'skyblue', label='low')
            
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('C:/Users/jhjh3/Desktop/jicode/data/seoul_copy.csv', encoding='utf-8')
data = csv.reader(f)
next(data)
high=[]
low=[]

for row in data:
    if row[5] != '' and row[-2] != '':
        date = row[2].split('-')
        if 1996 <= int(date[0]) :
            if date[1]=='03' :
                high.append(float(row[5]))
                low.append(float(row[-2]))

                       
plt.plot(high, 'pink')
plt.plot(low, 'sky blue')
plt.show()
```
