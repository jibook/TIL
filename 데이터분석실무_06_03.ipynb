```python
import pandas as pd
import numpy as np
```

```python
seoul_starbucks_df = pd.read_excel('../6_Starbucks_Location/files/seoul_starbucks_df.xlsx')
seoul_starbucks_df.head()
```

```python
seoul_starbucks = pd.read_excel('../6_Starbucks_Location/files/seoul_starbuck_final_list.xlsx')
seoul_starbucks.head()
```

```python
starbucks_sgg_count = seoul_starbucks.pivot_table(index = '시군구명', 
                                                values='매장명', 
                                                aggfunc='count'
                                                ).rename(columns={'매장명':'스타벅스_매장수'}, index={'시군구명':'시군구명'})
starbucks_sgg_count.head()
```

```python
seoul_sgg = pd.read_excel('../6_Starbucks_Location/files/seoul_sgg_list.xlsx')
seoul_sgg.head()
```

```python
seoul_sgg = pd.merge(seoul_sgg, starbucks_sgg_count, how='left', on='시군구명')
seoul_sgg.head()
```

```python
seoul_sgg_pop = pd.read_excel('../6_Starbucks_Location/files/sgg_pop.xlsx')
seoul_sgg_pop.head()
```

```python
seoul_sgg_biz = pd.read_excel('../6_Starbucks_Location/files/sgg_biz.xlsx')
seoul_sgg = pd.merge(
    seoul_sgg, 
    seoul_sgg_biz,
    how='left',
    on='시군구명'
)
seoul_sgg.head()
```

```python
seoul_sgg.to_excel('../6_Starbucks_Location/files/seoul_sgg_stat_class.xlsx', index = False)
```
