```python
# sample_1
code_master = pd.read_excel('./files/sample_codemaster.xlsx',
                          header = 0,
                          usecols = 'A:B',
                          skipfooter= 0)

code_master.head()
```

```python
sample_1_code = pd.merge(left = sample_1, 
                        right = code_master,
                        how = 'inner',   # outer
                        left_on = '국적코드',
                        right_on = '국적코드')

sample_1_code
```

```python
sample_1_code = pd.merge(left = sample_1, 
                        right = code_master,
                        how = 'left',
                        left_on = '국적코드',
                        right_on = '국적코드')

sample_1_code
```

```python
sample_2 = pd.read_excel('./files/sample_2.xlsx',
                        header = 1,
                        skipfooter=2,
                        usecols='A:C')
sample_2
```

```python
sample_2['기준년월'] = '2019-12'

sample_2_code = pd.merge(left = sample_2,
                        right = code_master,
                        how = 'left',
                        left_on='국적코드',
                        right_on='국적코드')

sample_2_code
```

```python
# sample_2_code['test'] = 'A'
sample = sample_1_code.append(sample_2_code,
                             ignore_index = False)

# sample
sample_concat = pd.concat([sample_1_code, sample_1_code],
                         ignore_index = True,
                         axis = 0)

sample_concat
```
