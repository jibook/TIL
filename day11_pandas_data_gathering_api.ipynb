```python
# https://weather.naver.com/today

html = requests.get('https://weather.naver.com/today/')
soup = bs(html.text, 'html.parser')
print(soup)
```

```python
dustdata_one = soup.find('ul',{'class':'today_chart_list'})
print(dustdata_one)
```

```python
dustdata_all = dustdata_one.findAll('div')
print(dustdata_all)
```
